```python
import numpy as np
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import torch

# Sample video data (your data here)

# Combine descriptions and keywords for training BERT embeddings
documents = [video['description'] + ' '.join(video['keywords']) for video in sample_videos]

# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sample query
query = "machine learning"

# Tokenize the search query and videos
tokenized_query = tokenizer(query, return_tensors='pt', padding=True, truncation=True)
tokenized_docs = tokenizer(documents, return_tensors='pt', padding=True, truncation=True)

# Get BERT embeddings for query and videos
with torch.no_grad():
    query_embeddings = model(**tokenized_query).last_hidden_state[:, 0, :].numpy()
    doc_embeddings = model(**tokenized_docs).last_hidden_state[:, 0, :].numpy()

# Normalize the embeddings
query_embeddings = query_embeddings / np.linalg.norm(query_embeddings, axis=1, keepdims=True)
doc_embeddings = doc_embeddings / np.linalg.norm(doc_embeddings, axis=1, keepdims=True)

# Calculate cosine similarities between query and videos
cosine_similarities = np.dot(query_embeddings, doc_embeddings.T).flatten()

# Sort and recommend top N videos
recommended_indices = cosine_similarities.argsort()[:-4:-1]  # Recommend top 3 videos

recommended_videos = []
for i in recommended_indices:
    video_key = sample_videos[i]['key']
    video_description = sample_videos[i]['description']
    recommended_videos.append({'key': video_key, 'description': video_description})

print("Recommended Videos for :", query)
for video in recommended_videos:
    print("Key:", video['key'])
    print("Description:", video['description'])
    print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Sample video data
sample_videos = sample_videos

# Combine descriptions and keywords for training TF-IDF
documents = [video['description'] + ' '.join(video['keywords']) for video in sample_videos]

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Apply k-means clustering
num_clusters = 5  # Choose the number of clusters (you can tune this)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_assignments = kmeans.fit_predict(tfidf_matrix)

def get_related_videos(query):
    # Vectorize the search query
    query_vector = tfidf_vectorizer.transform([query])

    # Transform query vector using the TF-IDF model
    query_tfidf = query_vector.toarray()

    # Find the cluster of the query
    query_cluster = kmeans.predict(query_tfidf)

    # Find videos in the same cluster as the query
    cluster_videos = [sample_videos[i] for i, cluster in enumerate(cluster_assignments) if cluster == query_cluster]

    # Calculate cosine similarities between query and cluster videos
    query_cluster_embeddings = tfidf_matrix[cluster_assignments == query_cluster]
    similarity_scores = cosine_similarity(query_tfidf, query_cluster_embeddings).flatten()

    # Sort and recommend top N videos
    recommended_indices = np.argsort(similarity_scores)[::-1][:3]
    recommended_videos = [
        {'key': cluster_videos[i]['key'], 'description': cluster_videos[i]['description']}
        for i in recommended_indices
    ]

    return recommended_videos

# Example query
query = "cake"
related_videos = get_related_videos(query)
for i in related_videos:
    print(i)
```
