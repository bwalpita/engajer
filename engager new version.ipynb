```python
import numpy as np
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import torch

# Sample video data (your data here)

# Combine descriptions and keywords for training BERT embeddings
documents = [video['description'] + ' '.join(video['keywords']) for video in sample_videos]

# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sample query
query = "machine learning"

# Tokenize the search query and videos
tokenized_query = tokenizer(query, return_tensors='pt', padding=True, truncation=True)
tokenized_docs = tokenizer(documents, return_tensors='pt', padding=True, truncation=True)

# Get BERT embeddings for query and videos
with torch.no_grad():
    query_embeddings = model(**tokenized_query).last_hidden_state[:, 0, :].numpy()
    doc_embeddings = model(**tokenized_docs).last_hidden_state[:, 0, :].numpy()

# Normalize the embeddings
query_embeddings = query_embeddings / np.linalg.norm(query_embeddings, axis=1, keepdims=True)
doc_embeddings = doc_embeddings / np.linalg.norm(doc_embeddings, axis=1, keepdims=True)

# Calculate cosine similarities between query and videos
cosine_similarities = np.dot(query_embeddings, doc_embeddings.T).flatten()

# Sort and recommend top N videos
recommended_indices = cosine_similarities.argsort()[:-4:-1]  # Recommend top 3 videos

recommended_videos = []
for i in recommended_indices:
    video_key = sample_videos[i]['key']
    video_description = sample_videos[i]['description']
    recommended_videos.append({'key': video_key, 'description': video_description})

print("Recommended Videos for :", query)
for video in recommended_videos:
    print("Key:", video['key'])
    print("Description:", video['description'])
    print()
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

data_dict = {
    "title": [
        "Introduction of Engajer",
        "Why Engajer",
        "Video Editing Methods",
        # ... (more titles)
    ],
    "link": [
        "https://engajer.com/introduction",
        "https://engajer.com/why",
        "https://engajer.com/editing",
        # ... (more links)
    ]
}

# Load pre-trained model and tokenizer
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Encode all titles
encoded_titles = [tokenizer(title, return_tensors="pt", padding=True) for title in data_dict['title']]

def get_related_title_and_link(question, encoded_titles, data_dict):
    input_question = tokenizer(question, return_tensors="pt", padding=True)
    scores = []
    for title_encoding in encoded_titles:
        with torch.no_grad():
            outputs = model(**title_encoding, **input_question)
        logits = outputs.logits
        scores.append(logits)
    best_match_idx = torch.argmax(torch.stack(scores)).item()
    related_title = data_dict['title'][best_match_idx]
    related_link = data_dict['link'][best_match_idx]
    return related_title, related_link

# Example usage
question = "How can I edit videos with Engajer?"
related_title, related_link = get_related_title_and_link(question, encoded_titles, data_dict)

if related_title is not None and related_link is not None:
    print(f"Related Title: {related_title}")
    print(f"Related Link: {related_link}")
else:
    print("No related information found.")
```
